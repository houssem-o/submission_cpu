```python
from explore_logs import *
```

#### Loading log data

```python
c10_npr = load_logs([f'nb201_cifar10_no_pretraining_{i}_logs.pickle' for i in range(10)])
c10_pr = load_logs([f'nb201_cifar10_pretrained_{i}_logs.pickle' for i in range(10)])

c100_npr = load_logs([f'nb201_cifar100_no_pretraining_{i}_logs.pickle' for i in range(10)])
c100_pr = load_logs([f'nb201_cifar100_pretrained_{i}_logs.pickle' for i in range(10)])

i120_npr = load_logs([f'nb201_ImageNet16-120_no_pretraining_{i}_logs.pickle' for i in range(10)])
i120_pr = load_logs([f'nb201_ImageNet16-120_pretrained_{i}_logs.pickle' for i in range(10)])
```

#### Table 1

```python
print('CIFAR 10')
print('No pretraining:')
print('Acc = {}\tRho = {}\tIdx = {}\tF = {}'.format(*grab_data(c10_npr)))
print('Pretrained')
print('Acc = {}\tRho = {}\tIdx = {}\tF = {}\n'.format(*grab_data(c10_pr)))

print('CIFAR 100')
print('No pretraining:')
print('Acc = {}\tRho = {}\tIdx = {}\tF = {}'.format(*grab_data(c100_npr)))
print('Pretrained')
print('Acc = {}\tRho = {}\tIdx = {}\tF = {}\n'.format(*grab_data(c100_pr)))

print('ImageNet16-120')
print('No pretraining:')
print('Acc = {}\tRho = {}\tIdx = {}\tF = {}'.format(*grab_data(i120_npr)))
print('Pretrained')
print('Acc = {}\tRho = {}\tIdx = {}\tF = {}\n\n\n'.format(*grab_data(i120_pr)))
```

#### Stats on indices

```python
print('CIFAR 10\t\tmin = {}\tmax = {}\tavg = {}\tstd = {}'.format(*get_index_stats(c10_pr)))
print('CIFAR 100\t\tmin = {}\tmax = {}\tavg = {}\tstd = {}'.format(*get_index_stats(c100_pr)))
print('ImageNet16-120\t\tmin = {}\tmax = {}\tavg = {}\tstd = {}\n\n'.format(*get_index_stats(i120_pr)))
```

#### Table 2

```python
snapshots = [100, 200, 400, 444]

print('CIFAR 10')
print('Not pretrained:')
for s in snapshots:
    print('Snapshot at {}\t\tn_evals = {}\t\tacc = {}'.format(s, *get_values_at_x(c10_npr, s)))
print('Pretrained:')
for s in snapshots:
    print('Snapshot at {}\t\tn_evals = {}\t\tacc = {}'.format(s, *get_values_at_x(c10_pr, s)))

print('\n\nCIFAR 100')
print('Not pretrained:')
for s in snapshots:
    print('Snapshot at {}\t\tn_evals = {}\t\tacc = {}'.format(s, *get_values_at_x(c100_npr, s)))
print('Pretrained:')
for s in snapshots:
    print('Snapshot at {}\t\tn_evals = {}\t\tacc = {}'.format(s, *get_values_at_x(c100_pr, s)))

print('\n\nImageNet16-120')
print('Not pretrained:')
for s in snapshots:
    print('Snapshot at {}\t\tn_evals = {}\t\tacc = {}'.format(s, *get_values_at_x(i120_npr, s)))
print('Pretrained:')
for s in snapshots:
    print('Snapshot at {}\t\tn_evals = {}\t\tacc = {}'.format(s, *get_values_at_x(i120_pr, s)))
```

#### Plots

```python
plt.rc('legend',fontsize=16)

with plt.style.context('ggplot'):
    plot_average_logs_multiple_experiments('CIFAR10',
                                           [c10_npr, c10_pr],
                                           ['MLP - No pretraining', 'MLP - pretrained'],
                                           bench=True, title=False)
    plot_average_logs_multiple_experiments('CIFAR100',
                                           [c100_npr, c100_pr],
                                           ['MLP - No pretraining', 'MLP - pretrained'],
                                           bench=True, title=False)
    plot_average_logs_multiple_experiments('ImageNet16-120',
                                           [i120_npr, i120_pr],
                                           ['MLP - No pretraining', 'MLP - pretrained'],
                                           bench=True, title=False)
plt.show()
```
